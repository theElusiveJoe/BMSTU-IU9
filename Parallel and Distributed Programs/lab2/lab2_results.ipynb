```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import os
import re
```

```python
def draw_file(f):
    strings = f.read().split('\n')
    title, *strings = strings
    data = np.array(list(map(lambda x: [int(x.split()[0]), float(x.split()[1])], filter(bool, strings))))
    data = data.T
    fig = plt.figure()
    ax1 = fig.add_axes((0.1,0.4,0.8,0.5))
    ax1.set_title(title)
    ax1.set_xlabel('количество процессов')
    ax1.set_ylabel('время(с.)')
    ax1.scatter(data[0],data[1], c='r')
    plt.show()
```

```python
files = os.listdir()
for file in files:
    if re.fullmatch('.*\.txt', file):
        with open(file, 'r') as f:
            draw_file(f)
```
